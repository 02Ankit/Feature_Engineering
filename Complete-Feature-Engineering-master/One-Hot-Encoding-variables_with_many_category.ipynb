## One Hot Encoding-Variables with many categories

```python
#https://www.kaggle.com/datasets/philanipro/mercedesbenz-greener-manufacturing
import pandas as pd
import numpy as np
#lets load the mercedes benz dataset for demonstration only the categorical variable
data = pd.read_csv('mercedesbenz.csv')
data
```

```python
data = pd.read_csv('mercedesbenz.csv', usecols = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
data
```

```python
#let's have a look at how many Labels each variables has 
for col in data.columns:
    print(col, ';', len(data[col].unique()),'labels')
```

```python
# lets examine how many columns we will obtain one hot encoding these variable
dummies  = pd.get_dummies(data, drop_first=True)
  
```

```python
dummies
```

We can observe that from with just 2 categorical features we are getting 69 features with the help of one hot encoding.

### KDD Cup Orange Challenge
What can we do instead?

http://proceedings.mlr.press/v7/niculescu09/niculescu09.pdf In the winning solution of the KDD 2009 cup: "Winning the KDD Cup Orange Challenge with Ensemble

The Team suggested using 10 most frequent labels convert them into dummy variables using onehotencoding

How can we do that in python?

```python
# lets find the top 10 most frequent categories for the variable x2
data.X1.value_counts().sort_values(ascending = False).head(30)
```

```python
# lets find the top 10 most frequent categories for the variable x2
dummies.X1_aa.value_counts().sort_values(ascending = False).head(20)
```

```python
# lets find the top 10 most frequent categories for the variable x2
data.X2.value_counts().sort_values(ascending = False).head(30)
```

```python
top10 = [x for x in data.X2.value_counts().sort_values(ascending = False).head(10).index]
top10
```

```python
# and now we make the 10 binary variable 
for label in top10:
    data[label] = np.where(data['X2']==label, 1,0)
data[['X2']+top10].head(40)
```

```python
#get whole set of dummy variables, for all categorical variables
def one_hot_top_x(df, variable, top_x_labels):
    #function to create the dummy variables for the frequent labels 
    #we can vary the number of most frequent tables that we encode
    for label in top_x_labels:
        df[variable+''+label] = np.where(data[variable]==label, 1,0)
#read the data again
data = pd.read_csv('mercedesbenz.csv', usecols = ['X1','X2','X3','X4','X5','X6'])
#encodeX2 into the 10 most frequent categories
one_hot_top_x(data, 'X2', top10)
data.head()
```

```python
# find the 10 mostfrequent categories for X1 
top10 = [x for x in data.X1.value_counts().sort_values(ascending = False).head(10).index]
#now ceate the 10 most frequent dummy variables for X1
one_hot_top_x(data, 'X1', top10)
data.head()
```
